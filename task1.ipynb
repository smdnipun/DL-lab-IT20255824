```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#  Create a random 4x4 array
random_array = np.random.exponential(scale=1.0, size=(4, 4))
print("Random 4x4 Array from Exponential Distribution:")
print(random_array)
print()

# Create a random 100000x1 array 

# Generate random arrays from exponential, uniform, and normal distributions
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.normal(loc=0.0, scale=1.0, size=100000)

# Plot histograms for all three distributions
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# Using Matplotlib, plot Z = X**2 + Y**2 in a 3D plot. 

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z = X**2 + Y**2')
plt.show()

# Task 1.4 - Calculate the Pearson standard coefficient and Spearman rank

file_path = "/content/drive/MyDrive/lab1/Pokemon.csv"
pokemon_data = pd.read_csv(file_path, encoding='latin1')
print(df.columns)

# Extracting the required features
required_features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
pokemon_required_data = pokemon_data[required_features]

# Calculate the Pearson standard coefficient
pearson_corr = pokemon_required_data.corr(method='pearson')

# Calculate the Spearman rank correlation
spearman_corr = pokemon_required_data.corr(method='spearman')

# Visualize the Pearson correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Pearson Correlation Heatmap')
plt.show()

# Visualize the Spearman correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Spearman Correlation Heatmap')
plt.show()
```
